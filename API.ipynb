**Chargement des prérequis depuis le Drive**

```
import joblib

# Model
model = joblib.load('model.pkl')

#Préprocessing
power_transform_scaler_object = joblib.load('power_transform_scaler_data.pkl')
min_max_scaler_object = joblib.load('min_max_scaler_data.pkl')
ind_features_to_pt = [0, 1, 7, 12, 13, 14, 16, 17, 21, 26, 27, 30, 35, 36, 38]
```

```
def preprocess(df,
               power_transform_scaler_object,
               min_max_scaler_object,
               ind_features_to_pt):
  #Power Transform Scaler
  features_to_pt = [df.columns[ind] for ind in ind_features_to_pt]
  df[features_to_pt] = power_transform_scaler_object.transform(df[features_to_pt])
  print("after pt")
  #Min Max Scaler
  features_to_mm = [f for f in df.columns if f not in features_to_pt]
  df[features_to_mm] = min_max_scaler_object.transform(df[features_to_mm])
  print("after mm")
  return df
```

```
from flask import Flask, request
from flask_ngrok import run_with_ngrok
import sklearn

app = Flask(__name__)
run_with_ngrok(app)

@app.route('/predict',
           methods = ['POST'])

def predict():
     df = pd.read_json(request.data)
     df = preprocess(df,
                     power_transform_scaler_object,
                     min_max_scaler_object,
                     ind_features_to_pt)
     prediction = model.predict(df)
     df_out = pd.DataFrame(prediction,
                           index = df.index,
                           columns = ["prediction (is_RB)"])
     json_out = df_out.to_json()
     return json_out
```

```
app.run()
```
